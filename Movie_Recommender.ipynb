```python
# importimg all the important libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import sklearn
import pickle 
```

```python
# reading the csv file of top movies 

data=pd.read_csv("top10K-TMDB-movies.csv")
```

From the following dataset we can see that we don't need the columns like "release_date" , "vote_average","vote_count" ,"original_language"


```python
# we can drop the columns which are not needed 

movie=data[["id" ,"title","genre","overview"]]
```

```python
movie.head()
```

Now we can merge the two columns i.e. genre and overview for further process named it as "review"

```python
movie["review"]=movie["genre"]+movie["overview"]
```

```python
movie.head()
```

Now I don't need these two columns anymore i.e. "genre" and "overview"

```python
movie.drop(["genre","overview"],axis=1,inplace=True)
```

```python
movie.head()
```

Now we are going to implement feature extraction on it we have to convert the text data of review into feature vector so that we can feed it on our model

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words="english")
```

```python
cv
```

```python
vector=cv.fit_transform(movie["review"].values.astype("U")).toarray()
```

```python
vector.shape
```

```python
vector
```

Now we hvae to find the similarity between the vectors and recommend the movies on the basis of similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
movie[movie["title"]=="The Shawshank Redemption"].index[0]
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])))
```

```python
distance = sorted(list(enumerate(similarity[0])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(movie.iloc[i[0]].title)
```

Now we'r gonna make a function named it as recommend and pass movie name so that it can search the related movies only

```python
def recommend(movie_name):
    index=movie[movie["title"]==movie_name].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(movie.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```

We will save the model for the further use using library in python named as pickle

```python
pickle.dump(movie,open("movie_list.pkl","wb"))

pickle.dump(similarity,open("similarity.pkl","wb"))
```
